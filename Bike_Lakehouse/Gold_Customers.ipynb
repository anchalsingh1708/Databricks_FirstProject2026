#Business Transformation and Modelling

```python
query= """
select
    row_number() over(order by ci.customer_id) as customer_key,
    ci.customer_id,
    ci.firstname,
    ci.lastname
    
from silver.crm_customers ci
"""

df=spark.sql(query)
```

#Write to Gold

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("gold.dim_customers")
)
```
